# Distribution of Numerical Features

```python
def distribution_of_numeric_feature(column):
  ax = sns.histplot(df[column], stat='density', kde=True, color = 'blue')
  feature = df[column]
  ax.axvline(feature.mean(), color='#ff033e', linestyle='dashed', linewidth=2)  # Red indicates mean
  ax.axvline(feature.median(), color='#00ffff', linestyle='dashed', linewidth=2) # Cyan indicates median
  plt.title(f'{column}')
```
